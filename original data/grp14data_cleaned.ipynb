```python
#Rename Columns
df = df.rename(columns={"Your Sex?": "Sex", 
                  "Your Matric (grade 12) Average/ GPA (in %)":"HS_GPA",
                   "What year were you in last year (2023) ?":"year_of_study",
                   "What faculty does your degree fall under?":"field",
                   "Your 2023 academic year average/GPA in % (Ignore if you are 2024 1st year student)":"current_GPA",
                   "Your Accommodation Status Last Year (2023)":"off_campus",
                   "Monthly Allowance in 2023":"monthly_allowance",
                   "Were you on scholarship/bursary in 2023?":"scholarship",
                   "Additional amount of studying (in hrs) per week":"study_hours",
                   "How often do you go out partying/socialising during the week? ": "partying_frequency",
                   "On a night out, how many alcoholic drinks do you consume?":"drinks_consumed",
                   "How many classes do you miss per week due to alcohol reasons, (i.e: being hungover or too tired?)":"classes_missed",
                   "How many modules have you failed thus far into your studies?":"modules_failed", 
                   "Are you currently in a romantic relationship?":"relationship", 
                   "Do your parents approve alcohol consumption?":"parents_approval", 
                   "How strong is your relationship with your parent/s?":"parent_relationship"})
df.columns
```

```python
#Drop Timestamp Column
df=df.drop(columns="Timestamp")
df.head()
```

```python
#exclude Postgraduate student rows from data
mask=df.year_of_study != "Postgraduate"

df1=df.loc[mask]
df1.head()
```

```python
df2=df1.copy()
df2['year_of_study']=df1.year_of_study.replace({"2nd Year":2,
                                             "1st Year":1,
                                             "3rd Year":3,
                                             "4th Year":4})

df2.year_of_study.value_counts()
```

```python
df2.year_of_study.info()
```

```python
df1.scholarship.value_counts()
```

```python
df2['scholarship']=df1.scholarship.replace({"Yes (NSFAS, etc...)": "Yes"})
df2.scholarship.value_counts()
```

```python
df2['off_campus']=df1.off_campus.replace({'Private accommodation/ stay with family/friends':"Yes",
                                        'Non-private accommodation ie. Res':"No"})
df2.off_campus.value_counts()
```
