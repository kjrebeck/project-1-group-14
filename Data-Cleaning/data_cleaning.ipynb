```python
import csv
import pandas as pd
import numpy as np
```

```python
#data
file= "../original_data/Stats survey.csv"

orig_df = pd.read_csv(file)

orig_df.head()
```

```python
orig_df.info()
```

```python
print(orig_df.shape)
```

```python
#Drop Timestamp Column
df=orig_df.drop(columns="Timestamp")
df.columns
```

```python
#replace columns -formatted for coding

new_col=['gender', 'hs_gpa', 'year_of_study','field', 'current_gpa','off_campus', 
         'monthly_allowance', 'scholarship', 'study_hours','partying_frequency', 'drinks_consumed', 'classes_missed',
         'modules_failed', 'relationship', 'parents_approval','parent_relationship']

df.columns=new_col
df.head(1)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
#drop nulls for
#sex, hs gpa, scholarship, study hrs, party freq, drinks, classes, mods, relationship, parent approve, parent, relationship

col_null_drop = ['gender', 'hs_gpa', 'scholarship', 'study_hours','partying_frequency', 'drinks_consumed', 'classes_missed',
       'modules_failed', 'relationship', 'parents_approval',
       'parent_relationship']

df1=df.dropna(subset=col_null_drop, how='any')
df1.isna().sum()
```

```python
#fill missing fields, may drop later

fill_nulls={'year_of_study': 0, 
         'field': 'Other',
         'current_gpa': 0,
         "off_campus":"N/A", 
         "monthly_allowance":"N/A"}

df2=df1.fillna(fill_nulls)
df2.info()
```

```python
print(df2.shape)
```

```python
#tidy
#used chat gbt to simplify
replacement_dicts = {'year_of_study': {"1st Year": 1, "2nd Year": 2, "3rd Year": 3, "4th Year": 4, "Postgraduate": 5},
         'scholarship': {"Yes (NSFAS, etc...)": "Yes"},
         'off_campus': {'Private accommodation/ stay with family/friends': "Yes", 'Non-private accommodation ie. Res': "No"}}

# Apply the replacements
for column, replacements in replacement_dicts.items():
    df2[column] = df2[column].replace(replacements)

# Ensure year_of_study is an integer
df2['year_of_study'] = df2['year_of_study'].astype(int)
df2.head()
```

```python
df2[['hs_gpa', 'current_gpa']] = df2[['hs_gpa', 'current_gpa']].round(1)
df2.head()
```

```python
df2.scholarship.value_counts()
```

```python
df2.off_campus.value_counts()
```

```python
#exclude N/A rows for off campus and monthly allowance
mask2=df2.off_campus != "N/A"
mask3=df2.monthly_allowance != "N/A"

df3=df2.loc[mask2].loc[mask3]
df3.sort_values(by='year_of_study', ascending=False)
df3.head()
```

```python
data_clean=df3

data_clean.info()
```

```python
# Push the remade DataFrame to a new CSV file
data_clean.to_csv("data_clean_351.csv", index=False)
```
